```python
#Dependcies
import pandas as pd
```

```python
# Make a reference to the file path
csv_path = "Resources/Homicide_Reports.csv"
```

```python
# Read the CSV into a Pandas DataFrame
datasource_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
datasource_df.head()
```

```python
#cleanup and drop extra columns not needed in dataset 
clean_data_df = datasource_df.drop(columns=["Agency Code", "Agency Name","Agency Type","Victim Ethnicity",
                                            "Perpetrator Ethnicity", "Victim Count", "Perpetrator Count","Record Source"])
clean_data_df.head()
```

```python
# Check to see if there are any values with similar values in "Relationship"
clean_data_df["Relationship"].value_counts()
```

```python
# Combining similar Family together
clean_data_df = clean_data_df.replace(
    {"Mother": "Family", "Father": "Family", "Brother": "Family", "Daughter": "Family", 
    "Son": "Family", "Sister": "Family", "In-Law": "Family"})
clean_data_df.head()
```

```python

# Combining similar datasets: Friend, Romantic Relationship, Step-family and Coworker together
clean_data_df = clean_data_df.replace(
    {"Acquaintance": "Friend", "Stepmother": "Step-Family", "Stepbrother": "Step-Family", "Stepdaughter": "Step-Family", 
    "Stepson": "Step-Family", "Stepsister": "Step-Family", "Stepfather": "Step-Family", "Wife": "Romantic", "Husband":
    "Romantic", "Girlfriend": "Romantic", "Boyfriend": "Romantic", "Ex-Wife": "Romantic", "Ex-Husband": "Romantic", 
    "Common-Law Wife": "Romantic", "Common-Law Husband": "Romantic", "Boyfriend/Girlfriend": "Romantic", "Employee": 
    "Coworker", "Employer": "Coworker"})
clean_data_df["Relationship"].value_counts()
```

```python
clean_data_df.head()
```

```python
print(clean_data_df["Weapon"].value_counts())
```

```python
# Combining similar datasets: Friend, Romantic Relationship, Step-family and Coworker together
clean_data_df = clean_data_df.replace(
    {"Handgun": "Firearm", "Shotgun": "Firearm", "Rifle": "Firearm", "Gun": "Firearm"})
```

```python
print(clean_data_df["Weapon"].value_counts())
```

```python
# Push the remade DataFrame to a new CSV file
clean_data_df.to_csv("Output/clean_data.csv",
                  encoding="utf-8", index=False, header=True)
```
