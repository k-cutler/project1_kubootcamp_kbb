```python
# Import Dependencies
import pandas as pd
```

```python
data1_csv = "Output/data1_clean.csv"
data2_csv = "Output/data2_clean.csv"
```

```python
data1_df = pd.read_csv(data1_csv)
data2_df = pd.read_csv(data2_csv)
```

```python
data1_df.head()
```

```python
data2_df.head()
```

```python
# Merge the two DataFrames together based on the State they share
crypto_df = pd.merge(data1_df, data2_df, on="State")
crypto_df.head(300)
```

```python
# Merge two dataframes using an outer join
crypto_df = pd.merge(data1_df, data2_df, on="State", how="outer")
crypto_df
```
