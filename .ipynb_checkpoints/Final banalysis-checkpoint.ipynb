```python
#Dependcies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reference file path
csv_path = "Output/clean_data.csv"

# Read the CSV into a Pandas DataFrame
homicide_reports_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
homicide_reports_df.head(800)
```

```python
print(homicide_reports_df["Weapon"].value_counts())
```

```python
# creating dataframe for weapons
weapon_df = homicide_reports_df.filter(['RecordID','Weapon','State','Perpetrator Sex'], axis=1)
weapon_df.head()
```

```python
# Using GroupBy in order to separate the data into fields according to "state" values
state_weapon_df = weapon_df.groupby(['State'])
state_weapon_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(weapon_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(weapon_df['Weapon'], weapon_df["Inches"], color='r', alpha=0.5, align="center")
plt.xticks(rotation="vertical")
plt.show()
```
